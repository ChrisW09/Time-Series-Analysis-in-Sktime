```python
import sktime
```

```python
from sktime import datasets
```

```python
airline = datasets.load_airline()
```

```python
airline
```

# Time series Data
## Univariate = One variable at a time for different time periods
## Multivariate = Mutiple variable at a time for different time

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(airline)
```

```python
from sktime.utils.plotting.forecasting import plot_ys
```

```python
plot_ys(airline[:30], airline[30:], labels=["First", "Second"])
```

```python
from sktime.forecasting.naive import NaiveForecaster
```

```python
from sktime.forecasting.model_selection import temporal_train_test_split
```

```python
y_train, y_test = temporal_train_test_split(airline, test_size=36)
```

```python
len(y_test)
```

```python
len(y_train)
```

```python
plot_ys(y_train, y_test, labels = ["train", "test"])
```

```python
forecaster = NaiveForecaster(strategy="last")
```

```python
forecaster.fit(y_train)
```

```python
import numpy as np
```

```python
fh = np.arange(1,len(y_test) + 1)
```

```python
fh
```

```python
y_pred = forecaster.predict(fh)
```

```python
plot_ys(y_train, y_test,y_pred, labels = ["train", "test", "pred"])
```
