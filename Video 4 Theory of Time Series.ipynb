```python
import numpy as np
```

```python
def smape(y_pred, y_true):
    num = np.abs(y_pred - y_true)
    deno = (np.abs(y_pred) + np.abs(y_true))/2
    result = num / deno
    return (100/len(result))*result.sum()
```
